```python
category = np.array(['A','B','C','D'])
v1 = np.array([3,8,1,10])
v2 = np.array([6,2,7,5])
v3 = np.array([4,1,8,4])
bar_width = 0.30
```

```python
# Pie Chart
#explode = (0,0,0.1,0)
cat = np.array(['A','B','C','D'])
val = np.array([3,5,8,10])
plt.pie(val, labels=cat,explode=(0.1,0.1,0.1,0.1),autopct='%1.1f%%')  # autopct for persentage , explode for 
plt.title("Pie Chart")
plt.show()
```

```python
# Donut Chart
#explode = (0,0,0.1,0)
cat = np.array(['A','B','C','D'])
val = np.array([3,5,8,10])
fig, ax = plt.subplots()
ax.pie(val, labels=cat,radius =1,startangle = 1)  # autopct for persentage , explode for 
ax.pie([1])
plt.title("Pie Chart")
plt.show()
```

```python
#### SKLearn ####
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset('iris')
df
```

```python
df['species'].value_counts()
```

```python
x=df.iloc[:,0:4].values
x
```

```python
y =df.iloc[:,4].values
y
```

```python
# Convert Y in numbers encoding 
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder_y = LabelEncoder()
```

```python
y =labelencoder_y.fit_transform(y)
y
```

```python
# For train test
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.3)
print(f'x_train',x_train)
print(f'x_test',x_test)
print(f'y_train',y_train)
print(f'y_test',y_test)
```

```python
x_train.size
```

```python
x_test.size
```

```python
y_train.size
```

```python
y_test.size
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
logmodel.fit(x_train,y_train)
```

```python
y_pred = logmodel.predict(x_test)
y_pred
```

```python
y_pred.sort()
y_pred
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

```python
logmodel.score(x_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
classifier_knn = KNeighborsClassifier(n_neighbors= 5,metric='minkowski',p=2)
```

```python
classifier_knn.fit(x_train,y_train)
```

```python
y_pred_knn =classifier_knn.predict(x_test)
y_pred_knn
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred_knn)
```

```python
classifier_knn.score(x_test,y_test)
```

```python
# NaiveBias Model 
from sklearn.naive_bayes import GaussianNB
```

```python
classifier_nb= GaussianNB()
```

```python
classifier_nb.fit(x_train,y_train)
```

```python
y_pred= classifier_nb.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred_knn,y_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
classifier_nb.score(x_test,y_test)
```
