```python
df = sns.load_dataset('iris')
df
```

```python
df.plot(kind= 'scatter',x ='petal_length',y ='sepal_length')
plt.grid()
plt.show()
```

```python
sns.set(style='darkgrid')
sns.scatterplot(x ='sepal_length',y ='petal_length',hue='species',data=df)
```

```python
sns.FaceGrid(df,hue='sepecies',height = 6).malt.scatter,'sepal_length','petal_length')
```

```python
sns.FacetGrid(df,hue='species',height=6).map(plt.scatter,'sepal_length','petal_length')
```
